Goal: To analyze the survey data of the interesting score about different data science project.

```python
# load the data
import pandas as pd

url="https://cocl.us/datascience_survey_data"
df=pd.read_csv(url,index_col=0)
df.head(6)
```

```python
interest = df.sort_values(by='Very interested',ascending=False)
interest = round(interest/ 2233,2)
```

Requirement:


To create this bar chart, you can follow the following steps:

Sort the dataframe in descending order of Very interested.
Convert the numbers into percentages of the total number of respondents. Recall that 2,233 respondents completed the survey. Round percentages to 2 decimal places.
As for the chart:
use a figure size of (20, 8),
bar width of 0.8,
use color #5cb85c for the Very interested bars, color #5bc0de for the Somewhat interested bars, and color #d9534f for the Not interested bars,
use font size 14 for the bar labels, percentages, and legend,
use font size 16 for the title, and,
display the percentages above the bars as shown above, and remove the left, top, and right borders.

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
width=0.8

fig, axes = plt.subplots(figsize=(20,8))
interest.plot(ax = axes, kind='bar',width=width,color=["#5cb85c","#5bc0de","#d9534f"],fontsize=14)
plt.legend(fontsize=14)
axes.set_title("Percentage of Respondents' Interest in Data Science Area",fontsize=16)

for p in axes.patches:
    axes.annotate(p.get_height(), (p.get_x()+p.get_width()/2, p.get_height()) ,
                  ha='center',va='center',
                  xytext=(0, 10), textcoords='offset points',fontsize=14)


```
